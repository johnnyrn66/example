```python
import pandas as pd

context=pd.read_csv("1001081_kmeans.csv",encoding='BIG5')
context.columns=['號碼', '緯度', '經度','上午投遞成功率','下午投遞成功率','重量','分類']
context.head()
```

```python
s=context[["緯度","經度"]].copy()
s.head()
```

```python
distance_node=[]
for i in range(s.shape[0]-1):
    for j in range(i+1,s.shape[0]):
        distance_node.append((i,j))
len(distance_node)    
```

```python
str(s["緯度"].iloc[0])+','+str(s["經度"].iloc[0])
```

```python
import googlemaps
from datetime import datetime


morn_distance = []
morn_duration = []

gmaps = googlemaps.Client(key= 'AIzaSyCxsLtJsAnTnLJ25vHiyc2w7HNPM-JRnhA' )
#now = datetime.now()
morn_time = datetime.strptime("2019-6-3 10:00:00", "%Y-%m-%d %H:%M:%S")


for i in distance_node:
    directions_result = gmaps.directions(str(s["緯度"].iloc[i[0]])+','+str(s["經度"].iloc[i[0]]),
                                     str(s["緯度"].iloc[i[1]])+','+str(s["經度"].iloc[i[1]]),
                                     mode="driving",
                                     avoid="tolls",
                                     departure_time=morn_time 
                                    )
    morn_distance.append(directions_result[0]['legs'][0]['distance']['text'])
    morn_duration.append(directions_result[0]['legs'][0]['duration']['text'])
    print(directions_result[0]['legs'][0]['distance']['text'])
    print(directions_result[0]['legs'][0]['duration']['text'])
```

```python
import csv
with open('1001081_ODpairs_morn.csv', 'w', newline='') as csvfile:
    # 以空白分隔欄位，建立 CSV 檔寫入器
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(['起點', '緯度', '經度','迄點', '緯度', '經度','距離','時間'])
    
    for j,i in enumerate(distance_node):
        writer.writerow([i[0], s["緯度"].iloc[i[0]], s["經度"].iloc[i[0]],i[1],s["緯度"].iloc[i[1]],s["經度"].iloc[i[1]],morn_distance[j],morn_duration[j]])

        
#臺北體育場郵局:南京東路3段346號後棟  25.051250, 121.548230    
```
